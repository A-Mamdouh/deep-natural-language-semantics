```python
import json
from pathlib import Path
from dataclasses import dataclass, asdict
from typing import List, Tuple
nl_path = Path("data3.txt")
ann_path = Path("annotations.json")
```

```python
dialogs: List[List[str]] = []
with open("data3.txt", "r") as fp:
    dialogs = fp.readlines()
    dialogs = [[s.strip() for s in d.split(".")] for d in dialogs]
```

```python
raw_annotations: dict = None
with open(ann_path, "r") as fp:
    raw_annotations = json.load(fp)
```

```python
@dataclass
class SentenceAnnotation:
    sentence: str
    subject: str
    verb: str
    object_: str
    adjectives: List[Tuple[str, str]]
    negated: bool = False
    always: bool = False


@dataclass
class DialogAnnotation:
    sentences: List[SentenceAnnotation]
```

```python
d_anns = []
for dialog, raw_d_ann in zip(dialogs, raw_annotations):
    anns = []
    for sentence, raw_ann in zip(dialog, raw_d_ann):
        anns.append(asdict(SentenceAnnotation(**{**raw_ann, "sentence": sentence})))
    d_anns.append(anns)
```

```python
with open("annotations2.json", "w") as fp:
    json.dump({"annotations": d_anns}, fp)
```
