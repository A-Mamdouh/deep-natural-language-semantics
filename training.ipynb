```python
def print_formulas(*formulas: Formula, sort: bool = True, indent: int = 0) -> None:
    indentation = "  " * indent
    sep = f"{indentation}- "
    if sort:
        formulas = sorted(map(str, formulas), key=lambda x: (len(x), x))
    print(sep, end="")
    print(*formulas, sep=f"\n{sep}")
```
